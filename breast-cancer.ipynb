# EDA

```python
#Distribution graph
plt.figure(figsize=(20,15))
plotnumber=1
for column in df:
    #if plotnumber<=30:
        ax = plt.subplot(6,6, plotnumber)
        sns.kdeplot(df[column])
        plt.xlabel(column)
    #else:
    #   break
        plotnumber+=1
plt.tight_layout()
plt.show()
```

```python
#By the graph i found that mostly it was right skewed and se is very small 
```

```python
plt.figure(figsize=(20,12))
corr=df.corr()
mask = np.triu(np.ones_like(corr, dtype=bool))
sns.heatmap(corr,mask=mask, linewidths=1, annot=True, fmt = ".2f")
plt.show()
```

```python
#There is multicollinearity in the dataset we can remove those features which are highly correlated>0.9 
```

```python
df.shape[1]
```

```python
df.drop('id',axis=1,inplace=True)
```

```python
corr_matrix = df.corr().abs()
mask = np.triu(np.ones_like(corr_matrix, dtype=bool))
tri_df = corr_matrix.mask(mask)
drop_Col = [x for x in tri_df.columns if any(tri_df[x]>0.92)]
df = df.drop(drop_Col, axis=1)
print(df.shape[1])
```
